Для начала обсудим условия решения задачи:

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
import re
                           
def creating_work_exp_months_column(rec):
    if rec == 'Не указано' or rec is np.nan:
        return np.nan
    
    pattern = re.compile(r'\d+\s[гл]|\d+\sм')
    mch = pattern.findall(str(rec)[:30])
    total = 0
    for i in mch:
        if 'л' in i or 'г' in i:
            i = i.split()
            total += int(i[0]) * 12
        if 'м' in i:
            i = i.split()
            total += int(i[0])
    return int(total)

data['Опыт работы (месяц)'] = data['Опыт работы'].apply(creating_work_exp_months_column)
data.drop('Опыт работы', axis=1, inplace=True)
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
def creating_city_removal_trip_columns(rec):
    million_cities = ['Новосибирск', 'Екатеринбург', 'Нижний Новгород', 'Казань', 'Челябинск', 'Омск', 'Самара', 'Ростов-на-Дону',
                      'Уфа', 'Красноярск', 'Пермь', 'Воронеж', 'Волгоград' ]
    # Приводим к списку
    rec_list = rec.split(',')
    if len(rec_list) == 4:
        del rec_list[1]
    for i in range(len(rec_list)):
        rec_list[i] = rec_list[i].strip()
    # Выделяем город
    if rec_list[0] == 'Москва':
        city = 'Москва'
    elif rec_list[0] == 'Санкт-Петербург':
        city = 'Санкт-Петербург'
    elif rec_list[0] in million_cities:
        city = 'город-миллионник'
    else:
        city = 'другие'
    # Выделяем готовность к переезду
    if 'не ' in rec_list[1]:
        removal = False
    else:
        removal = True
    # Выделяем готовность к командировкам
    if 'не ' in rec_list[2]:
        trip = False
    else:
        trip = True
    # Возвращаем значения
    return city, removal, trip

data['Город'], data['Готовность к переезду'], data['Готовность к командировкам'] = zip(*data['Город, переезд, командировки'].apply(creating_city_removal_trip_columns))
data['Город'] = data['Город'].astype('category')
data.drop('Город, переезд, командировки', axis=1, inplace=True)
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
def encoding_job_time(rec):
    job_time = ['полная занятость', 'частичная занятость', 'проектная работа', 'волонтерство', 'стажировка']
    flags = []
    # Приводим к списку
    rec_list = rec.split(',')
    for i in range(len(rec_list)):
        rec_list[i] = rec_list[i].strip()
    # Выделяем флаги
    for i in job_time:
        if i in rec_list:
            flags.append(True)
        else:
            flags.append(False)
    return flags

def encoding_schedule(rec):
    schedule = ['полный день', 'сменный график', 'гибкий график', 'удаленная работа', 'вахтовый метод']
    flags = []
    # Приводим к списку
    rec_list = rec.split(',')
    for i in range(len(rec_list)):
        rec_list[i] = rec_list[i].strip()
    # Выделяем флаги
    for i in schedule:
        if i in rec_list:
            flags.append(True)
        else:
            flags.append(False)
    return flags

data['полная занятость'], data['частичная занятость'], data['проектная работа'], data['волонтерство'], data['стажировка'] = zip(*data['Занятость'].apply(encoding_job_time))
data['полный день'], data['сменный график'], data['гибкий график'], data['удаленная работа'], data['вахтовый метод'] = zip(*data['График'].apply(encoding_schedule))
data.drop(['Занятость', 'График'], axis=1, inplace=True)
```

```python
exchange_data = pd.read_csv('D:\\Datasets\\ExchangeRates.csv')
exchange_data['date'] = pd.to_datetime(exchange_data['date'], dayfirst=True)

data['Обновление резюме'] = pd.to_datetime(pd.to_datetime(data['Обновление резюме'], dayfirst=True).dt.date)

def currency_renaming(rec):
    rec = rec.split()
    for i in range(len(rec)):
        rec[i] = rec[i].strip()
    if rec[1] == 'бел.руб.':
        rec[1] = 'BYN'
    elif rec[1] == 'грн.':
        rec[1] = 'UAH'
    elif rec[1] == 'сум':
        rec[1] = 'UZS'
    return int(rec[0]), rec[1]

data['Сумма ЗП'], data['Валюта'] = zip(*data['ЗП'].apply(currency_renaming))
data.drop('ЗП', axis=1, inplace=True)

data = data.merge(exchange_data[['currency', 'date', 'close', 'proportion']], left_on=['Обновление резюме', 'Валюта'], right_on=['date', 'currency'], how='left')
data[['close', 'proportion']] = data[['close', 'proportion']].fillna(1, axis=1)
data['ЗП (руб)'] = data['Сумма ЗП'] * data['close'] / data['proportion']
data.drop(['Сумма ЗП', 'Валюта', 'currency', 'date', 'close', 'proportion'], axis=1, inplace=True)
```

```python
# Получается безобразный график. Очистим график методом Тьюки.

def outliers_iqr(data, feature):
    x = data[feature]
    quartile_1, quartile_3 = x.quantile(0.25), x.quantile(0.75),
    iqr = quartile_3 - quartile_1
    lower_bound = quartile_1 - (iqr * 1.5)
    upper_bound = quartile_3 + (iqr * 1.5)
    outliers = data[(x<lower_bound) | (x > upper_bound)]
    cleaned = data[(x>lower_bound) & (x < upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_iqr(data, 'ЗП (руб)')
```

```python
def outliers_z_score(data, feature, log_scale=False):
    if log_scale:
        x = np.log(data[feature])
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - 3 * sigma
    upper_bound = mu + 4 * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x > lower_bound) & (x < upper_bound)]
    return outliers, cleaned

outliers, cleaned = outliers_z_score(dedupped, 'Возраст', log_scale=True)
dedupped = dedupped.drop(outliers.index, axis=0)
outliers
```
